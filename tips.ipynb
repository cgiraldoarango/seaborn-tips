# Uso de seaborn
## Dataset tips

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

datos = sns.load_dataset("tips")
df = pd.DataFrame(datos)
print(df.head())

# grafico de dispersion entre size y tip
sns.scatterplot(data=df, x="size", y="tip")
plt.xlabel("SIZE")
plt.ylabel("TIP")
plt.title("Gráfico de dispersión")
plt.show()
```

## Gráfico de barras

```python
# grafico de barras propinas promedio por tamaña fiesta
sns.barplot(data=df, x="size", y="tip", errorbar=None)
plt.show()
```

### Agregar una columna al dataframe

```python
#agregar una columna
#que calcule el porcentaje de propinas
df["porcentaje"]=(df["tip"]/df["total_bill"]) *100

print(df)
```
